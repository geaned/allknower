```python
import pandas as pd
from sklearn.model_selection import StratifiedGroupKFold
```

```python
df = pd.read_csv("./wiki_ir_processed.csv")
df.head()
```

```python
splitter = StratifiedGroupKFold(n_splits=5, shuffle=True)
```

```python
next(splitter.split(df, df["label"], groups=df["query"]))
```

```python
train_index, val_index = next(splitter.split(df, df["label"], groups=df["query"]))
train_data, val_data = df.loc[train_index], df.loc[val_index]
```
